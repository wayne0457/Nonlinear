```python
import numpy as np
import matplotlib.pyplot as plt

# r = a exp (k * theta)
def spiral(round, pitch_angle, theta, a):
    k = np.tan(pitch_angle)
    return a * np.exp(k * theta)

round = 2
pitch_angle = np.radians(10)
count = 1000
theta = np.linspace(0.0, round * 2 * np.pi, count)
a = 1
mag = 20 * np.exp(-theta/100)


r1 = spiral(round, pitch_angle, theta, a)
x1 = r1 * np.cos(theta)
y1 = r1 * np.sin(theta)

theta2 = theta + np.radians(120)
r2 = spiral(round, pitch_angle, theta, a)
x2 = r2 * np.cos(theta2)
y2 = r2 * np.sin(theta2)

theta3 = theta + np.radians(240)
r3 = spiral(round, pitch_angle, theta, a)
x3 = r3 * np.cos(theta3)
y3 = r3 * np.sin(theta3)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
```

```python
from scipy.interpolate import griddata

x = np.linspace(np.min(x1), np.max(x1), 1000)
y = np.linspace(np.min(y1), np.max(y1), 1000)
grid_x, grid_y = np.meshgrid(x,y)
#plt.plot(grid_x, grid_y, '.')
#plt.plot(grid_x, grid_y, '.')
points1 = np.array([x1, y1]).T
points2 = np.array([x2, y2]).T
points3 = np.array([x3, y3]).T
points = np.concatenate((points1,points2, points3), axis=0)

mag3 = np.concatenate((mag, mag, mag))
grid_r = griddata(points, mag3, (grid_x, grid_y), method='nearest', fill_value=0)
#print(grid_r)
plt.imshow(grid_r, extent=(x[0],x[-1],y[0],y[-1]), origin='lower')
```

The pitch angle  of a spiral

Logarithmic spirals are characterized by the property that the pitch angle remains invariant for all points of the spiral. For an Archimedean spiral the angle decreases with the distance, while for a hyperbolic spiral the angle increases with the distance.


In general, logarithmic spirals are good approximations of the shape of galactic spiral arms (Seigar & James 1998).
(https://iopscience.iop.org/article/10.1088/0067-0049/199/2/33/pdf)
